## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell
The Wizard Inventroy Program

COMMAND MENU
walk - Walk down the path 
show - Show all items 
drop - Drop an item
exit - Exit program

Command: walk
While walking down a path, you see a scroll of uncursing. 
Do you want to grab it? (y/n): y
You picked up a scroll of uncursing.

Command: walk
While walking down a path, you see an unknown potion. 
Do you want to grab it? (y/n): y
You can't carry any more items. Drop something first.

Command: show
1.	a wooden staff
2.	a scroll of invisibility
3.	a crossbow
4.	a scroll of uncursing

Command: drop Number: 3
You dropped a crossbow.

Command: exit 
Bye!
```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying (`wizard_inventory.txt`). Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry **four** items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
### CODE HERE ###
import random

def display_menu():
    print("The Wizard Inventory Program\n")
    print("COMMAND MENU")
    print("walk - Walk down the path")
    print("show - Show all items")
    print("drop - Drop an item")
    print("exit - Exit program\n")

def read_inventory():
    """Reads the current inventory from the file if it exists."""
    inventory = []
    try:
        with open("wizard_inventory.txt", "r") as file:
            for line in file:
                inventory.append(line.strip())
    except FileNotFoundError:
        # If the file doesn't exist yet, it will be created when they grab an item.
        pass
    return inventory

def write_inventory(inventory):
    """Writes the current inventory list to the file."""
    with open("wizard_inventory.txt", "w") as file:
        for item in inventory:
            file.write(item + "\n")

def walk(inventory):
    """Reads all possible items, picks one randomly, and handles grabbing."""
    try:
        with open("wizard_all_items.txt", "r") as file:
            all_items = [line.strip() for line in file]
    except FileNotFoundError:
        print("Error: Could not find 'wizard_all_items.txt' in the current directory.\n")
        return

    # Randomly pick an item
    item = random.choice(all_items)
    print(f"While walking down a path, you see {item}.")
    
    choice = input("Do you want to grab it? (y/n): ").lower().strip()
    
    if choice == 'y':
        if len(inventory) >= 4:
            print("You can't carry any more items. Drop something first.\n")
        else:
            inventory.append(item)
            write_inventory(inventory)
            print(f"You picked up {item}.\n")
    else:
        print()

def show(inventory):
    """Displays the current inventory."""
    if len(inventory) == 0:
        print("Your inventory is empty.\n")
    else:
        for i, item in enumerate(inventory, start=1):
            print(f"{i}. {item}")
        print()

def drop(inventory):
    """Handles dropping an item and updates the file."""
    if len(inventory) == 0:
        print("Your inventory is empty. There is nothing to drop.\n")
        return

    try:
        number = int(input("Number: "))
        if number < 1 or number > len(inventory):
            print("Invalid item number.\n")
        else:
            # Remove item
            dropped_item = inventory.pop(number - 1)
            write_inventory(inventory)
            print(f"You dropped {dropped_item}.\n")
    except ValueError:
        print("Invalid input. Please enter a valid number.\n")

def main():
    display_menu()
    inventory = read_inventory()

    while True:
        command = input("Command: ").lower().strip()
        
        if command == "walk":
            walk(inventory)
        elif command == "show":
            show(inventory)
        elif command == "drop":
            drop(inventory)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.\n")

if __name__ == "__main__":
    main()
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesn’t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
### CODE HERE ###
import csv

# Constant for the filename so it's easy to change if needed
FILENAME = "monthly_sales.csv"

def display_menu():
    print("Monthly Sales program\n")
    print("COMMAND MENU")
    print("monthly - View monthly sales")
    print("yearly  - View yearly summary")
    print("edit    - Edit sales for a month")
    print("exit    - Exit program\n")

def read_sales():
    """Reads the sales from the CSV file and returns a list of lists."""
    sales = []
    try:
        with open(FILENAME, mode="r", newline="") as file:
            reader = csv.reader(file)
            for row in reader:
                # Append the month and the sales amount
                sales.append([row[0], int(row[1])])
    except FileNotFoundError:
        print(f"Error: Could not find '{FILENAME}' in the current directory.\n")
    return sales

def write_sales(sales):
    """Writes the updated sales list back to the CSV file."""
    with open(FILENAME, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerows(sales)

def view_monthly(sales):
    """Displays all 12 months and their sales."""
    for row in sales:
        month = row[0]
        amount = row[1]
        print(f"{month} - {amount}")
    print()

def view_yearly(sales):
    """Calculates and displays the yearly total and monthly average."""
    # Calculate the total using a list comprehension
    total = sum([row[1] for row in sales])
    
    # Calculate the average and round to a max of 2 decimal places
    count = len(sales)
    average = round(total / count, 2) if count > 0 else 0
    
    print(f"Yearly total:\t{total}")
    print(f"Monthly average:\t{average}\n")

def edit_sales(sales):
    """Allows the user to edit a month's sales and saves immediately."""
    # Create a list of valid months (e.g., ['Jan', 'Feb', 'Mar'...]) to check against
    valid_months = [row[0].title() for row in sales]
    
    # Get user input and format it to match the list (e.g., 'jan' becomes 'Jan')
    month_input = input("Three-letter Month: ").title()
    
    if month_input not in valid_months:
        print("Invalid three-letter abbreviation.\n")
        return

    try:
        new_amount = int(input("Sales Amount: "))
    except ValueError:
        print("Invalid input. Please enter a valid number.\n")
        return

    # Find the month in the list and update the amount
    for row in sales:
        if row[0].title() == month_input:
            row[1] = new_amount
            break

    # Immediately save the updated list to the CSV file
    write_sales(sales)
    print(f"Sales amount for {month_input} was modified.\n")

def main():
    display_menu()
    sales = read_sales()

    # If the file wasn't found, exit the program early to avoid crashing
    if not sales:
        return

    while True:
        command = input("Command: ").lower().strip()
        
        if command == "monthly":
            view_monthly(sales)
        elif command == "yearly":
            view_yearly(sales)
        elif command == "edit":
            edit_sales(sales)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command. Please try again.\n")

if __name__ == "__main__":
    main()
````
