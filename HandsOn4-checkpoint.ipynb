# Prime Number Checker

```python
def get_valid_integer():
    # Loop until the user gives us a valid number
    while True:
        number = int(input("Please enter an integer between 1 and 5000: "))
        if number >= 1 and number <= 5000:
            return number
        else:
            print("Error: Number must be between 1 and 5000. Try again.\n")

def get_factors(number):
    
    factors = []
    
    for i in range(1, number + 1):
        if number % i == 0:
        
            factors.append(i)
            
    return factors

def main():
    print("Prime Number Checker\n")
    
    keep_going = "y"
    while keep_going.lower() == "y":
        
        
        number = get_valid_integer()
        
        
        factors = get_factors(number)
        
        
        if len(factors) == 2:
            print(str(number) + " is a prime number.\n")
        else:
            print(str(number) + " is NOT a prime number.")
            
            
            factors_string = ""
            for f in factors:
                factors_string += str(f) + " "
                
            print("It has " + str(len(factors)) + " factors: " + factors_string + "\n")
            
        keep_going = input("Try again? (y/n): ")
        print() 
        
    print("Bye!")


main()
```

# Contact Manager

```python
def display_title():
    print("Contact Manager\n")

def display_menu():
    print("COMMAND MENU")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add  - Add a contact")
    print("del  - Delete a contact")
    print("exit - Exit program")
    print()

def list_contacts(contacts):
    if len(contacts) == 0:
        print("No contacts available.\n")
    else:
        for i in range(len(contacts)):
            print(str(i + 1) + ". " + contacts[i][0])
        print()

def view_contact(contacts):
    number = int(input("Number: "))
    
    if number < 1 or number > len(contacts):
        print("Error: Invalid contact number.\n")
    else:
        contact = contacts[number - 1]
        print("Name: " + contact[0])
        print("Email: " + contact[1])
        print("Phone: " + contact[2] + "\n")

def add_contact(contacts):
    name = input("Name: ")
    email = input("Email: ")
    phone = input("Phone: ")
    
    contacts.append([name, email, phone])
    print(name + " was added.\n")

def delete_contact(contacts):
    number = int(input("Number: "))
    
    if number < 1 or number > len(contacts):
        print("Error: Invalid contact number.\n")
    else:
        contact = contacts.pop(number - 1)
        print(contact[0] + " was deleted.\n")

def main():

    contacts = [
        ["Guido van Rossum", "guido@python.org", "555-555-5555"],
        ["Eric Idle", "eric@ericidle.com", "+44 20 7946 0958"]
    ]
    
    display_title()
    display_menu()
    
    while True:
        command = input("Command: ")
        print()
        
        if command == "list":
            list_contacts(contacts)
        elif command == "view":
            view_contact(contacts)
        elif command == "add":
            add_contact(contacts)
        elif command == "del":
            delete_contact(contacts)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Not a valid command. Please try again.\n")


main()
```
