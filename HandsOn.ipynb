# Even Or Odd Checker

```python
def check_even_odd(number):
    if number % 2 == 0:
        return "even"
    else:
        return "odd"

def main():
    print("Even or Odd Checker\n")
    
    number = int(input("Enter an integer: "))
    
    result = check_even_odd(number)
    
    print(str(number) + " is " + result + ".")

main()
```

# Feet and Meters Converter

```python
import Conversions

def display_title():
    print("Feet and Meters Converter\n")

def display_menu():
    print("Conversions Menu:")
    print("a. Feet to Meters")
    print("b. Meters to Feet")

def main():
    display_title()
    
    keep_going = "y"
    while keep_going.lower() == "y":
        display_menu()
        
        choice = input("Select a conversion (a/b): ")
        print() 
        
        if choice == "a":
            feet = float(input("Enter feet: "))
            meters = Conversions.to_meters(feet)
            print(str(meters) + " meters\n")
            
        elif choice == "b":
            meters = float(input("Enter meters: "))
            feet = Conversions.to_feet(meters)
            print(str(feet) + " feet\n")
            
        else:
            print("Invalid choice. Please enter a or b.\n")
            
        keep_going = input("Would you like to perform another conversion? (y/n): ")
        print() 
        
    print("Thanks, bye!")

main()
```

# Sales Tax Calculator

```python
import Sales_Tax

def display_title():
    print("Sales Tax Calculator\n")

def main():
    display_title()
    
    keep_going = "y"
    while keep_going.lower() == "y":
        # Get input
        total = float(input("Enter total: "))
        
        
        tax = Sales_Tax.get_tax(total)
        grand_total = Sales_Tax.get_total_after_tax(total)
        
        
        print("Total:           " + str(round(total, 2)))
        print("Sales tax:       " + str(tax))
        print("Total after tax: " + str(grand_total) + "\n")
        
        # Ask to continue
        keep_going = input("Enter another? (y/n): ")
        print() 
        
    print("Goodbye!")

main()
```

# Prime Number Checker

```python
def get_valid_integer():
    
    number = int(input("Please enter an integer: "))
    return number

def get_factor_count(number):
    
    count = 0
    for i in range(1, number + 1):
        if number % i == 0:
            count += 1
    return count

def main():
    print("Prime Number Checker\n")
    
    keep_going = "y"
    while keep_going.lower() == "y":
        
        
        number = get_valid_integer()
        
        
        factors = get_factor_count(number)
        
        
        if factors == 2:
            print(str(number) + " is a prime number.\n")
        else:
            print(str(number) + " is NOT a prime number.")
            print("It has " + str(factors) + " factors.\n")
            
        keep_going = input("Try again? (y/n): ")
        print() 
        
    print("Goodbye!")

main()
```
